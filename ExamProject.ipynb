## Data Preparation

```python
#Split data into target (y) and features (X)
y = data['quality']
data = data.drop('quality', axis=1)
X = data

y2 = data2['quality']
data2 = data2.drop('quality', axis=1)
X2 = data2

data3 = data3.drop('quality', axis=1)
X3 = data3
```

```python
from sklearn.model_selection import train_test_split
```

```python
#split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2)
```

```python
#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier

#Create Random Forest Classifier
clf=RandomForestClassifier(n_estimators=1000, max_depth=14, oob_score=True)

clf2=RandomForestClassifier(n_estimators=50, max_depth=10, oob_score=True)

#Put our data through the Random Forest Classifier
clf.fit(X_train, y_train)

clf2.fit(X2_train, y2_train)
```

```python
#Test accuracy agains the train data
clf.score(X_train, y_train)
```

```python
clf2.score(X2_train, y2_train)
```

```python
#Test accuracy agains the test data
clf.score(X_test, y_test)
```

```python
clf2.score(X2_test, y2_test)
```
